### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.490142
longitude = 0.08384

distance = {'RatingValue':'5','geocode.latitude': 
                         {'$gte': (latitude - degree_search), 
                          '$lt': (latitude + degree_search)
                         },'geocode.longitude':
                         {'$gte': (longitude - degree_search), 
                          '$lt': (longitude + degree_search)
                         }
    }

fields = {'LocalAuthorityBusinessID':1,
          'BusinessName':1,
          'BusinessType':1,
          'RatingValue': 1, 
          'geocode.latitude': 1,
          'geocode.longitude': 1,
          'scores.Hygiene': 1,
          'LocalAuthorityName': 1   
}

## Sort query in ascending order
sort =  [('scores.Hygiene', 1)]

## Limit results to five
limit = 5

## Combine queries
results = list(artifacts.find(distance, fields).sort(sort).limit(limit))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
distance_from_rest = pd.DataFrame(results)
```

```python
distance_from_rest.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results

hygine_match = {'$match': {'scores.Hygiene': 0}}

local_author = {'$group': {'_id': "$LocalAuthorityName",'count_per_local': {'$sum':1}}}

sort = {'$sort': {'count_per_local': -1}}

pipeline = [hygine_match, local_author, sort]
results = list(artifacts.aggregate(pipeline))
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results_DF = pd.DataFrame(results)
```

```python
# Display the number of rows in the DataFrame
len(results_DF)
```
